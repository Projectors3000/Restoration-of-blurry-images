```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df_err = pd.read_csv('rmse.csv')
df_loss = pd.read_csv('loss.csv')
```

```python
df_err_z = pd.read_csv('rmse_zeros.csv')
df_loss_z = pd.read_csv('loss_zeros.csv')
df_err_r = pd.read_csv('rmse_random.csv')
df_loss_r = pd.read_csv('loss_random.csv')
```

```python
plt.figure(figsize = (12, 4))
plt.suptitle('MSE for different initializations')


plt.subplot(121)
plt.semilogy(df_err['0'][:50], label='Blurred initialization')
plt.semilogy(df_err_z['0'][:50], label='Initialization with zeros')
plt.semilogy(df_err_r['0'][:50], label='Random initialization')
plt.title('Loss with l1 norm')
plt.ylabel('log(MSE)')
plt.xlabel('iteration number')
plt.legend()


plt.subplot(122)
plt.semilogy(df_err['1'][:50], label='Blurred initialization')
plt.semilogy(df_err_z['1'][:50], label='Initialization with zeros')
plt.semilogy(df_err_r['1'][:50], label='Random initialization')
plt.title('Loss with l2 norm')
plt.ylabel('log(MSE)')
plt.xlabel('iteration number')
plt.legend()
plt.savefig('MSE_prox.png')
plt.show()
```

```python
plt.figure(figsize = (12, 4))
plt.suptitle('Loss for different initializations')

plt.subplot(121)
plt.semilogy(df_loss['0'][:50], label='Blurred initialization')
plt.semilogy(df_loss_z['0'][:50], label='Initialization with zeros')
plt.semilogy(df_loss_r['0'][:50], label='Random initialization')
plt.title('Loss with l1 norm')
plt.ylabel('log(loss)')
plt.xlabel('iteration number')
plt.legend()


plt.subplot(122)
plt.semilogy(df_loss['1'][:50], label='Blurred initialization')
plt.semilogy(df_loss_z['1'][:50], label='Initialization with zeros')
plt.semilogy(df_loss_r['1'][:50], label='Random initialization')
plt.title('Loss with l1 norm')
plt.ylabel('log(loss)')
plt.xlabel('iteration number')
plt.legend()

plt.savefig('loss_prox.png')
plt.show()
#plt.savefig('loss_prox.png')
```
